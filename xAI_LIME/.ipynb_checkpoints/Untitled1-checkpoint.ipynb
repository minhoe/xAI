### 이미지 분류 예시 (1)
- 어떠한 부분이 특정 객체로 인식하는데 긍정적 영향(녹색)인지 부정적 영향(빨간색)인지 제시
- 아래 그림은 동물의 얼굴을 보고 객체를 제대로 분류했음을 확인할 수 있음



### 이미지 분류 예시 (2)
- 우측 테니스공은 테니스공을 제대로 인식하고 객체를 올바르게 판단함
- 좌측은 '둥근모양+공 위의 선'을 보고 판단하는 야구공의 특성에 따라 테니스공도 일부 야구공으로 분류 가능함을 확인
   - 야구공이 갖는 특성을 보다 잘 이해하도록 추가 학습이 필요!



### 이미지 분류 예시 (3)
- 좌측은 포크 모양을 참고하였음에도'주걱'으로 잘못 분류하였음
   - '주걱'과 '포크'의 차이를 보다 명확하게 분류할 수 있는 이미지가 추가로 필요함을 확인
