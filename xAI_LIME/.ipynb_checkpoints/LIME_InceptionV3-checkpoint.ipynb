```python
import os
import keras
from keras.applications import inception_resnet_v2
from keras.applications.inception_resnet_v2 import InceptionResNetV2
from keras.applications.inception_resnet_v2 import preprocess_input
from keras.applications.inception_resnet_v2 import decode_predictions

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras import backend as K
from skimage.io import imread

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
print('Notebook run using keras:', keras.__version__)
```

### Using InceptionResNetV2 Model as a object detection model

```python
# create the base pre-trained model
base_model = InceptionResNetV2(weights='imagenet', include_top=True)
#base_model = VGG19(weights='imagenet', include_top=True)
```

```python
def transform_img_fn(path_list):
    output = []
    for img_path in path_list:
        img = load_img(img_path, target_size=(299,299)) # For Inception_resnet_v2
        #img = load_img(img_path, target_size=(224,224)) # For VGG19
        
        x = img_to_array(img)  # this is a Numpy array with shape (3(RGB), width, height)
        x = np.expand_dims(x, axis=0) # this is a Numpy array with shape (1, 3(RGB), width, height)
        x = preprocess_input(x)
        output.append(x)
    return np.vstack(output)
```

### Sample images

```python
import glob

file_list = glob.glob("./Images/*/*", recursive=True)
print('Loading Images')
#print('Number of Images found : ', len(file_list))


file_list = [file_list[x] for x in [0,3,5]] 
img_list = transform_img_fn(file_list)
#img_list = [img_list[x] for x in [0,3,5]]

fig = plt.figure()
fig.set_size_inches(10,10)

for idx, img in enumerate(img_list):
    ax = plt.subplot(1,3,idx+1)
    ax.set_title(file_list[idx].split('/')[2])
    plt.axis('off')
    plt.imshow(img_list[idx] / 2 + 0.5)
```

### For sample images, do predict what the object is by A.I.

```python
#preds = base_model.predict(np.expand_dims(img_list[0], axis=0))
preds = base_model.predict(img_list, verbose=0)

result = decode_predictions(preds, top=5)

fig = plt.figure()
fig.set_size_inches(16,10)

for idx, img in enumerate(img_list):
    ax = plt.subplot(1,3,idx+1)
    ax.set_title(file_list[idx].split('/')[2])
    plt.axis('off')
    plt.imshow(img_list[idx] / 2 + 0.5)

plt.show()
print('  %-25s %-25s %-25s' %(result[0][0][1]+ '(' + str(result[0][0][2].round(3)) + ')',
                        result[1][0][1]+ '(' + str(result[1][0][2].round(3)) + ')',
                        result[2][0][1]+ '(' + str(result[2][0][2].round(3))))

print('  %-25s %-25s %-25s' %(result[0][1][1]+ '(' + str(result[0][1][2].round(3)) + ')',
                        result[1][1][1]+ '(' + str(result[1][1][2].round(3)) + ')',
                        result[2][1][1]+ '(' + str(result[2][1][2].round(3)) + ')'))

print('  %-25s %-25s %-25s' %(result[0][2][1]+ '(' + str(result[0][2][2].round(3)) + ')',
                        result[1][2][1]+ '(' + str(result[1][2][2].round(3)) + ')',
                        result[2][2][1]+ '(' + str(result[2][2][2].round(3)) + ')'))
```
